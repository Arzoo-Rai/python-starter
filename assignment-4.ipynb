```python
import pandas as pd 
import numpy as np
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
indx = np.where(df['X'] == 0)
outindex = list()
for i in df.index:
    indexes = [abs(x-i) for x in indx]
    outindex.append(abs(min(indexes[0])))
df["y"] = outindex

df
    
```

```python
import pandas as pd
import numpy as np
import random as rn
bdays =pd.bdate_range('2015-01-01', '2015-12-31')
series = pd.DataFrame({"nm":[rn.random() for i in range(0,len(bdays))]},index=bdays)
series['days'] = series.index.day
series['date'] = pd.to_datetime(series.index)
series['daysname'] = series['date'].dt.day
series['monthname'] = series['date'].dt.month
# series['qtr'] = series['date'].dt.quarter
series2 = series.groupby(['monthname']).max()
# series2
series2.groupby(np.arange(len(series2.index))//4,axis=0).max()
```
